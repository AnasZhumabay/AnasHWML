Тапсырма 1

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect = Rectangle(5, 3)
print("Аудан:", rect.area())
print("Периметр:", rect.perimeter())
```

Тапсырма 2


```python
try:
    a = float(input("Бірінші санды енгізіңіз: "))
    b = float(input("Екінші санды енгізіңіз: "))
    result = a / b
except ZeroDivisionError:
    print("Қате: 0-ге бөлуге болмайды")
except ValueError:
    print("Қате: Сан енгізу керек")
else:
    print(f"{a} / {b} = {result}")
```
