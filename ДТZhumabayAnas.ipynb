```python
grades = {
    "Алия": 85,
    "Бекзат": 92,
    "Меруерт": 78,
    "Нұрлан": 95,
    "Дина": 88
}

max_grade = max(grades.values())

top_students = [name for name, grade in grades.items() if grade == max_grade]

print("Ең жоғары баға:", max_grade)
print("Ең жоғары бааны алған студент(тер):", top_students)
```
