Тапсырма 1: CSV Файлын Оқу және Тізімге Сақтау


```python
import csv

data_rows = []
with open('Salaries.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    header = next(reader)
    for i, row in enumerate(reader):
        if i < 10:
            data_rows.append(row)
        else:
            break

print("Data rows length:", len(data_rows))
print("First element:", data_rows[0])
```

Тапсырма 2: Кортеж және Сөздік – Алғашқы 3 Адамның Атын Сақтау


```python
first_three_names = tuple(row[0] for row in data_rows[:3])

employee_titles = {row[0]: row[2] for row in data_rows[:3]}

print("First three names (tuple):", first_three_names)
print("Employee titles (dict):", employee_titles)
```

Тапсырма 3: Тізімдік Өрнектер (List Comprehension)


```python
base_pays_str = [row[3] for row in data_rows]

print("Base pays (as strings):", base_pays_str)
```

Тапсырма 4: Жиын (Set) – Бірегей Атақтарды Табу


```python
job_titles_list = []
for row in data_rows:
    job_titles_list.append(row[2])

unique_titles = set(job_titles_list)

print("Unique job titles (set):", unique_titles)
```
